```python
import torch

size = 128
white_noise = torch.randn((size, size))
```

```python
import matplotlib.pyplot as plt
plt.imshow(white_noise, cmap='seismic')
```

```python
import scipy

gaussian_blob = scipy.ndimage.gaussian_filter(white_noise, sigma=5)
```

```python
plt.imshow(gaussian_blob, cmap='seismic')
```

```python
def cosine_correlation(rf, l, dx):
    for i in range(dx, rf.shape[0]-dx):
        for j in range(dx, rf.shape[1]-dx):
            rf[i, j] = rf[i,j] * torch.cos(torch.Tensor([2 * 3.1415 * (dx/l)]))
    return rf
```

```python
u_cos = cosine_correlation(white_noise, 10, 5)
```

```python
plt.imshow(u_cos, cmap='seismic')
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
J = 5
lj = 5

nx = 128
ny = 128

L = 2
x = np.linspace(-1, 1, nx)
y = np.linspace(-1, 1, ny)
[xx, yy] = np.meshgrid(x, y)
u = np.zeros((nx, ny))

for i in range(J):
    A = np.random.uniform(-0.5, 0.5)
    Kx = 2*np.pi*np.random.randint(1, lj)/L
    Ky = 2*np.pi*np.random.randint(1, lj)/L
    phi = np.random.uniform(0, 2*np.pi)

    u = u + A*np.sin(Kx * xx + Ky * yy + phi)
plt.imshow(u, cmap='seismic')
```
